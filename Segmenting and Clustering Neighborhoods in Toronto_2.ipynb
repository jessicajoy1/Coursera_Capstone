```python
from bs4 import BeautifulSoup
import requests
import pandas as pd

# download url data from internet
url = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
toronto_df = pd.read_html(url)[0]

toronto_df.head()
```

```python
# remove rows where Borough is 'Not assigned'
toronto_df2 = toronto_df[~ toronto_df['Borough'].str.contains('Not assigned')]
toronto_df2.reset_index(drop=True, inplace=True)
toronto_df2.head(10)
```

```python
# assign Neighbourhood=Borough where Neighbourhood is 'Not assigned'
toronto_df2['Neighbourhood'].replace('Not assigned', toronto_df2['Borough'])
toronto_df2.head(10)
```

```python
# group multiple Neighbourhood under one Postcode
toronto_df3 = toronto_df2.groupby(by=['Postcode','Borough'], as_index = False).aggregate(lambda x:','.join(x))
toronto_df3.head(10)
```
